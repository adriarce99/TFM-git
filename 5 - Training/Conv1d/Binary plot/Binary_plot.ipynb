```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
list_val = os.listdir('val')
list_train = os.listdir('train')
```

```python
df_val_list = []
for val_file in list_val:
    df = pd.read_csv("val/{}".format(val_file))
    df_val_list.append(df)
```

```python
df_train_list = []
for train_file in list_train:
    df = pd.read_csv("train/{}".format(train_file))
    df_train_list.append(df)
```

```python
%matplotlib inline
for val_file in list_val:
    df = pd.read_csv("val/{}".format(val_file))
    plt.plot(df['Step'], df['Value'], label = val_file)
plt.legend()
plt.xlim(0,51)
plt.ylabel("Binary Accuracy")
plt.xlabel("Epoch")
plt.savefig("Binary_all")
plt.show()
```

```python
%matplotlib inline

df_val = pd.read_csv("val/Conv1D_D_validation.csv")
plt.plot(df_val['Step'], df_val['Value'], label = "Conv1D_D_validation.csv")
df_train = pd.read_csv("train/Conv1D_D_train.csv")
plt.plot(df_train['Step'], df_train['Value'], label = "Conv1D_D_train.csv")
plt.legend(loc='lower right')
plt.xlim(0,51)
plt.ylabel("Binary Accuracy")
plt.xlabel("Epoch")
plt.savefig("Binary_best")
plt.show()
```
